```python
"""Perform RFM (Recency, Frequency, Monetary) Analysis"""
# Calculate Recency, Frequency, and Monetary values
today = full_data['TransactionDate'].max()
rfm = full_data.groupby('CustomerID').agg({
        'TransactionDate': lambda x: (today - x.max()).days,  # Recency
        'TransactionID': 'count',  # Frequency
        'TotalValue': 'sum'  # Monetary
    })
rfm.columns = ['Recency', 'Frequency', 'Monetary']
# Create RFM scores
rfm['R_Score'] = pd.qcut(rfm['Recency'], q=5, labels=[5, 4, 3, 2, 1])
rfm['F_Score'] = pd.qcut(rfm['Frequency'], q=5, labels=[1, 2, 3, 4, 5])
rfm['M_Score'] = pd.qcut(rfm['Monetary'], q=5, labels=[1, 2, 3, 4, 5])
# Calculate RFM Score
rfm['RFM_Score'] = rfm['R_Score'].astype(str) + rfm['F_Score'].astype(str) + rfm['M_Score'].astype(str)
# Segment customers
def segment_customers(row):
        if row['R_Score'] == 5 and row['F_Score'] == 5:
            return 'Champions'
        elif row['R_Score'] == 5:
            return 'Active'
        elif row['R_Score'] == 1:
            return 'Lost'
        elif row['F_Score'] >= 4:
            return 'Loyal'
        else:
            return 'Average'
    
rfm['Customer_Segment'] = rfm.apply(segment_customers, axis=1)
# Plot RFM segments
plt.figure(figsize=(10, 6))
segment_counts = rfm['Customer_Segment'].value_counts()
segment_counts.plot(kind='bar')
plt.title('Customer Segments Distribution')
plt.xlabel('Segment')
plt.ylabel('Number of Customers')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
def analyze_purchase_patterns(full_data):
    """Analyze customer purchase patterns and product combinations"""
    
    # Average time between purchases per customer
    customer_purchases = full_data.groupby('CustomerID').agg({
        'TransactionDate': lambda x: x.diff().mean()
    })
    print("\nAverage days between purchases:")
    print(customer_purchases['TransactionDate'].dt.days.describe())
    
    # Basket analysis - products bought together
    transactions_products = full_data.groupby('TransactionID')['ProductName'].agg(list)
    print("\nTop product combinations:")
    from collections import Counter
    combinations = Counter()
    for products in transactions_products:
        if len(products) > 1:
            # Get all possible pairs of products
            pairs = [(a, b) for idx, a in enumerate(products) 
                    for b in products[idx + 1:]]
            combinations.update(pairs)
    print("\nMost common product pairs:")
    for pair, count in combinations.most_common(5):
        print(f"{pair}: {count} times")

def analyze_price_sensitivity(full_data):
    """Analyze how price affects purchase behavior"""
    
    # Calculate price per unit
    full_data['PricePerUnit'] = full_data['TotalValue'] / full_data['Quantity']
    
    # Group by product and analyze price variations
    price_analysis = full_data.groupby('ProductID').agg({
        'PricePerUnit': ['mean', 'std', 'min', 'max'],
        'Quantity': 'sum'
    })
    
    # Correlation between price and quantity
    correlations = full_data.groupby('ProductID').apply(
        lambda x: x['PricePerUnit'].corr(x['Quantity'])
    )
    print("\nPrice-Quantity Correlation by Product:")
    print(correlations.describe())

def analyze_customer_segments(full_data):
    """Advanced customer segmentation based on multiple metrics"""
    
    # Calculate customer metrics
    customer_metrics = full_data.groupby('CustomerID').agg({
        'TotalValue': ['sum', 'mean'],
        'TransactionID': 'count',
        'Quantity': 'sum'
    })
    
    customer_metrics.columns = ['Total_Spent', 'Avg_Order_Value', 
                              'Purchase_Frequency', 'Total_Items']
    
    # Add derived metrics
    customer_metrics['Avg_Items_Per_Order'] = (
        customer_metrics['Total_Items'] / customer_metrics['Purchase_Frequency']
    )
    
    # Segment customers based on metrics
    from sklearn.preprocessing import StandardScaler
    from sklearn.cluster import KMeans
    
    # Prepare data for clustering
    features = ['Total_Spent', 'Purchase_Frequency', 'Avg_Order_Value']
    X = StandardScaler().fit_transform(customer_metrics[features])
    
    # Apply clustering
    kmeans = KMeans(n_clusters=4, random_state=42)
    customer_metrics['Segment'] = kmeans.fit_predict(X)
    
    # Analyze segments
    segment_analysis = customer_metrics.groupby('Segment').mean()
    print("\nCustomer Segments Analysis:")
    print(segment_analysis)

def analyze_promotional_effectiveness(full_data):
    """Analyze effectiveness of different price points"""
    
    # Calculate actual price per unit
    full_data['ActualPricePerUnit'] = full_data['TotalValue'] / full_data['Quantity']
    
    # Group by product to analyze price variations
    price_analysis = full_data.groupby(['ProductID', 'ProductName']).agg({
        'ActualPricePerUnit': ['mean', 'min', 'max'],
        'Quantity': 'sum',
        'TotalValue': 'sum'
    }).round(2)
    
    # Calculate price ranges and their effectiveness
    full_data['PriceRange'] = pd.qcut(full_data['ActualPricePerUnit'], 
                                     q=5, 
                                     labels=['Very Low', 'Low', 'Medium', 'High', 'Very High'])
    
    # Analyze metrics by price range
    price_range_analysis = full_data.groupby('PriceRange').agg({
        'Quantity': 'mean',
        'TotalValue': 'mean',
        'TransactionID': 'count'
    }).round(2)
    
    print("\nPrice Range Analysis:")
    print(price_range_analysis)
    
    # Analyze volume vs price relationship
    volume_price_corr = full_data.groupby('ProductID').apply(
        lambda x: x['ActualPricePerUnit'].corr(x['Quantity'])
    )
    
    print("\nPrice-Volume Correlation by Product:")
    print(volume_price_corr.describe().round(3))
    
    return price_range_analysis
def analyze_regional_performance(full_data):
    """Detailed regional performance analysis"""
    
    # Regional metrics
    regional_metrics = full_data.groupby('Region').agg({
        'TotalValue': ['sum', 'mean'],
        'CustomerID': 'nunique',
        'TransactionID': 'count'
    })
    
    # Calculate region-specific metrics
    regional_metrics['Orders_Per_Customer'] = (
        regional_metrics[('TransactionID', 'count')] / 
        regional_metrics[('CustomerID', 'nunique')]
    )
    
    print("\nRegional Performance Metrics:")
    print(regional_metrics)

def identify_key_insights(full_data):
    """Extract key business insights"""
    
    insights = []
    
    # Customer value insights
    top_customers = full_data.groupby('CustomerID')['TotalValue'].sum().nlargest(10)
    insights.append(f"Top 10 customers contribute ${top_customers.sum():,.2f} in revenue")
    
    # Product insights
    product_revenue = full_data.groupby('ProductName')['TotalValue'].sum()
    top_products = product_revenue.nlargest(5)
    insights.append(f"Top 5 products account for {(top_products.sum()/product_revenue.sum()*100):.1f}% of revenue")
    
    # Regional insights
    region_customers = full_data.groupby('Region')['CustomerID'].nunique()
    insights.append(f"Customer distribution: {region_customers.to_dict()}")
    
    # Growth insights
    monthly_growth = full_data.groupby(full_data['TransactionDate'].dt.to_period('M'))['TotalValue'].sum().pct_change()
    avg_monthly_growth = monthly_growth.mean() * 100
    insights.append(f"Average monthly growth rate: {avg_monthly_growth:.1f}%")
    
    print("\nKey Business Insights:")
    for i, insight in enumerate(insights, 1):
        print(f"{i}. {insight}")

# Run all analyses
analyze_purchase_patterns(full_data)
analyze_price_sensitivity(full_data)
analyze_customer_segments(full_data)
analyze_promotional_effectiveness(full_data)
analyze_regional_performance(full_data)
identify_key_insights(full_data)
```

```python
import pandas as pd
import numpy as np
import faiss
from sklearn.preprocessing import MinMaxScaler
from datetime import datetime

def create_balanced_features(customers_df, transactions_df, products_df):
    """Create balanced feature vectors for customer similarity"""
    
    # Convert dates
    customers_df['SignupDate'] = pd.to_datetime(customers_df['SignupDate'])
    transactions_df['TransactionDate'] = pd.to_datetime(transactions_df['TransactionDate'])
    current_date = transactions_df['TransactionDate'].max()
    
    # 1. RFM Score (40% weight)
    rfm = transactions_df.groupby('CustomerID').agg({
        'TransactionDate': lambda x: (current_date - x.max()).days,  # Recency
        'TransactionID': 'count',                                    # Frequency
        'TotalValue': 'sum'                                         # Monetary
    })
    rfm.columns = ['Recency', 'Frequency', 'Monetary']
    
    # Convert to scores (1-5) and explicitly convert to numeric
    rfm['RecencyScore'] = pd.qcut(rfm['Recency'].rank(method='first'), 
                                 q=5, labels=[5, 4, 3, 2, 1]).astype(float)
    rfm['FrequencyScore'] = pd.qcut(rfm['Frequency'].rank(method='first'), 
                                   q=5, labels=[1, 2, 3, 4, 5]).astype(float)
    rfm['MonetaryScore'] = pd.qcut(rfm['Monetary'].rank(method='first'), 
                                  q=5, labels=[1, 2, 3, 4, 5]).astype(float)
    
    # 2. Order Statistics (20% weight)
    order_stats = transactions_df.groupby('CustomerID').agg({
        'TotalValue': ['mean', 'std'],
        'Quantity': ['mean', 'std']
    }).fillna(0)
    order_stats.columns = ['AvgOrderValue', 'StdOrderValue', 
                          'AvgQuantity', 'StdQuantity']
    
    # 3. Category Preferences (30% weight)
    enriched_transactions = transactions_df.merge(products_df, on='ProductID')
    category_pivot = pd.crosstab(
        enriched_transactions['CustomerID'], 
        enriched_transactions['Category'],
        values=enriched_transactions['TotalValue'],
        aggfunc='sum',
        normalize='index'
    ).fillna(0)
    
    # 4. Purchase Timing (10% weight)
    timing_features = transactions_df.groupby('CustomerID').agg({
        'TransactionDate': [
            lambda x: x.dt.hour.mean(),     # Average purchase hour
            lambda x: x.dt.dayofweek.mean(), # Average day of week
            lambda x: x.dt.dayofweek.std(),  # Variation in purchase day
            lambda x: x.diff().dt.days.mean() # Average days between purchases
        ]
    }).fillna(0)
    timing_features.columns = ['AvgHour', 'AvgDayOfWeek', 
                             'DayVariation', 'PurchaseInterval']
    
    # Multiply features by their weights
    rfm_weighted = rfm[['RecencyScore', 'FrequencyScore', 'MonetaryScore']].multiply(0.4)
    order_stats_weighted = order_stats.multiply(0.2)
    category_pivot_weighted = category_pivot.multiply(0.3)
    timing_features_weighted = timing_features.multiply(0.1)
    
    # Combine all features
    customer_features = pd.concat([
        rfm_weighted,
        order_stats_weighted,
        category_pivot_weighted,
        timing_features_weighted
    ], axis=1).fillna(0)
    
    return customer_features

def find_balanced_lookalikes(customer_features, target_customers, k=3):
    """Find lookalike customers using FAISS"""
    
    # Scale features
    scaler = MinMaxScaler()
    normalized_features = scaler.fit_transform(customer_features.values.astype('float32'))
    normalized_features = np.ascontiguousarray(normalized_features)
    
    # Build FAISS index
    dimension = normalized_features.shape[1]
    faiss.normalize_L2(normalized_features)
    index = faiss.IndexFlatIP(dimension)
    index.add(normalized_features)
    
    customer_ids = customer_features.index.tolist()
    results = {}
    
    for target_id in target_customers:
        try:
            target_idx = customer_ids.index(target_id)
            query_vector = np.ascontiguousarray(normalized_features[target_idx:target_idx+1])
            
            # Find similar customers
            D, I = index.search(query_vector, k + 1)
            
            # Process results
            recommendations = []
            for sim, idx in zip(D[0], I[0]):
                if customer_ids[idx] != target_id:
                    # Transform similarity score
                    similarity_score = 1 / (1 + np.exp(-5 * (sim - 0.5)))
                    recommendations.append(f"{customer_ids[idx]}:{similarity_score:.3f}")
                if len(recommendations) == k:
                    break
                    
            results[target_id] = recommendations
            
        except ValueError:
            print(f"Warning: Customer {target_id} not found in dataset")
            results[target_id] = []
            
    return results

def main():
    """Main execution function"""
    
    # Create customer features
    print("Creating balanced customer features...")
    customer_features = create_balanced_features(customers_df, transactions_df, products_df)
    
    # Generate target customers (C0001-C0020)
    target_customers = [f'C{str(i).zfill(4)}' for i in range(1, 21)]
    
    # Find lookalikes
    print("Finding lookalike customers...")
    results = find_balanced_lookalikes(customer_features, target_customers)
    
    # Create output DataFrame
    final_df = pd.DataFrame({
        'cust_id': list(results.keys()),
        'lookalikes': [','.join(v) for v in results.values()]
    })
    
    # Save to CSV
    output_filename = 'FirstName_LastName_Lookalike.csv'
    final_df.to_csv(output_filename, index=False)
    print(f"\nResults saved to {output_filename}")
    
    # Display results
    print("\nLookalike Results for Customers C0001-C0020:")
    print(final_df.to_string())
    
    return final_df

# Run the model
if __name__ == "__main__":
    recommendations_df = main()
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from scipy.spatial.distance import cdist

def create_customer_features(customers_df, transactions_df, products_df):
    """Create customer feature vectors"""
    
    # Convert dates
    customers_df['SignupDate'] = pd.to_datetime(customers_df['SignupDate'])
    transactions_df['TransactionDate'] = pd.to_datetime(transactions_df['TransactionDate'])
    
    # 1. RFM Features
    current_date = transactions_df['TransactionDate'].max()
    
    rfm = transactions_df.groupby('CustomerID').agg({
        'TransactionDate': lambda x: (current_date - x.max()).days,  # Recency
        'TransactionID': 'count',                                    # Frequency
        'TotalValue': ['sum', 'mean']                               # Monetary
    })
    
    # Flatten column names
    rfm.columns = ['Recency', 'Frequency', 'TotalSpend', 'AvgSpend']
    
    # 2. Product Category Preferences
    # Merge transactions with products to get categories
    tx_with_products = transactions_df.merge(products_df, on='ProductID')
    
    # Calculate category preferences
    category_preferences = pd.crosstab(
        tx_with_products['CustomerID'], 
        tx_with_products['Category'],
        values=tx_with_products['TotalValue'],
        aggfunc='sum',
        normalize='index'
    ).fillna(0)
    
    # 3. Combine Features
    customer_features = rfm.join(category_preferences).fillna(0)
    
    return customer_features

def find_similar_customers(customer_features, target_ids, n_recommendations=3):
    """Find similar customers using cosine similarity"""
    
    # Standardize features
    scaler = StandardScaler()
    scaled_features = scaler.fit_transform(customer_features)
    
    # Calculate cosine similarity
    similarity_matrix = 1 - cdist(scaled_features, scaled_features, metric='cosine')
    
    # Create DataFrame for easier lookup
    similarity_df = pd.DataFrame(
        similarity_matrix,
        index=customer_features.index,
        columns=customer_features.index
    )
    
    results = {}
    for target_id in target_ids:
        # Get similarities for target customer
        customer_similarities = similarity_df[target_id].sort_values(ascending=False)
        
        # Remove self and get top N
        similar_customers = customer_similarities[customer_similarities.index != target_id][:n_recommendations]
        
        # Format results
        recommendations = [f"{idx}:{score:.3f}" for idx, score in similar_customers.items()]
        results[target_id] = recommendations
    
    return results

def main():
    """Main execution function"""
    print("Creating customer features...")
    customer_features = create_customer_features(customers_df, transactions_df, products_df)
    
    # Generate target customer IDs (C0001-C0020)
    target_customers = [f'C{str(i).zfill(4)}' for i in range(1, 21)]
    
    print("Finding similar customers...")
    results = find_similar_customers(customer_features, target_customers)
    
    # Create output DataFrame
    final_df = pd.DataFrame({
        'cust_id': list(results.keys()),
        'lookalikes': [','.join(v) for v in results.values()]
    })
    
    # Save results
    output_filename = 'Classic_FirstName_LastName_Lookalike.csv'
    final_df.to_csv(output_filename, index=False)
    print(f"\nResults saved to {output_filename}")
    
    # Display results
    print("\nLookalike Results for Customers C0001-C0020:")
    print(final_df.to_string())
    
    return final_df

# Run the model
if __name__ == "__main__":
    recommendations_df = main()
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import davies_bouldin_score
import matplotlib.pyplot as plt
import seaborn as sns

def create_clustering_features(customers_df, transactions_df, products_df):
    """Create features for clustering"""
    
    # Convert dates
    customers_df['SignupDate'] = pd.to_datetime(customers_df['SignupDate'])
    transactions_df['TransactionDate'] = pd.to_datetime(transactions_df['TransactionDate'])
    current_date = transactions_df['TransactionDate'].max()
    
    # 1. RFM Features
    rfm = transactions_df.groupby('CustomerID').agg({
        'TransactionDate': lambda x: (current_date - x.max()).days,  # Recency
        'TransactionID': 'count',                                    # Frequency
        'TotalValue': ['sum', 'mean', 'std']                        # Monetary
    })
    
    # Flatten column names
    rfm.columns = ['Recency', 'Frequency', 'TotalSpend', 'AvgSpend', 'StdSpend']
    
    # 2. Product Category Preferences
    tx_with_products = transactions_df.merge(products_df, on='ProductID')
    category_preferences = pd.crosstab(
        tx_with_products['CustomerID'], 
        tx_with_products['Category'],
        values=tx_with_products['TotalValue'],
        aggfunc='sum',
        normalize='index'
    ).fillna(0)
    
    # 3. Purchase Pattern Features
    purchase_patterns = transactions_df.groupby('CustomerID').agg({
        'Quantity': ['mean', 'std'],
        'TransactionDate': lambda x: x.dt.hour.mean()  # Average purchase hour
    })
    purchase_patterns.columns = ['AvgQuantity', 'StdQuantity', 'AvgPurchaseHour']
    
    # Combine all features
    features = rfm.join([category_preferences, purchase_patterns]).fillna(0)
    
    return features

def find_optimal_clusters(features, max_clusters=10):
    """Find optimal number of clusters using DB Index"""
    
    # Scale features
    scaler = StandardScaler()
    scaled_features = scaler.fit_transform(features)
    
    # Calculate DB Index for different numbers of clusters
    db_scores = []
    for n_clusters in range(2, max_clusters + 1):
        kmeans = KMeans(n_clusters=n_clusters, random_state=42)
        cluster_labels = kmeans.fit_predict(scaled_features)
        db_score = davies_bouldin_score(scaled_features, cluster_labels)
        db_scores.append(db_score)
        print(f"Clusters: {n_clusters}, DB Index: {db_score:.3f}")
    
    # Plot DB Index scores
    plt.figure(figsize=(10, 6))
    plt.plot(range(2, max_clusters + 1), db_scores, marker='o')
    plt.xlabel('Number of Clusters')
    plt.ylabel('Davies-Bouldin Index')
    plt.title('Davies-Bouldin Index by Number of Clusters')
    plt.grid(True)
    plt.show()
    
    # Find optimal number of clusters (minimum DB Index)
    optimal_clusters = db_scores.index(min(db_scores)) + 2
    return optimal_clusters, db_scores

def perform_clustering(features, n_clusters):
    """Perform clustering with the optimal number of clusters"""
    
    # Scale features
    scaler = StandardScaler()
    scaled_features = scaler.fit_transform(features)
    
    # Perform K-means clustering
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    cluster_labels = kmeans.fit_predict(scaled_features)
    
    # Add cluster labels to features
    clustered_data = features.copy()
    clustered_data['Cluster'] = cluster_labels
    
    return clustered_data, kmeans.cluster_centers_

def analyze_clusters(clustered_data):
    """Analyze and visualize clustering results"""
    
    # 1. Cluster sizes
    print("\nCluster Sizes:")
    print(clustered_data['Cluster'].value_counts().sort_index())
    
    # 2. Cluster profiles
    cluster_profiles = clustered_data.groupby('Cluster').mean()
    print("\nCluster Profiles:")
    print(cluster_profiles)
    
    # 3. Visualizations
    
    # PCA for dimensionality reduction
    from sklearn.decomposition import PCA
    pca = PCA(n_components=2)
    pca_features = pca.fit_transform(StandardScaler().fit_transform(
        clustered_data.drop('Cluster', axis=1)
    ))
    
    # Plot clusters
    plt.figure(figsize=(12, 8))
    scatter = plt.scatter(pca_features[:, 0], pca_features[:, 1], 
                         c=clustered_data['Cluster'], cmap='viridis')
    plt.xlabel('First Principal Component')
    plt.ylabel('Second Principal Component')
    plt.title('Customer Segments Visualization')
    plt.colorbar(scatter, label='Cluster')
    plt.show()
    
    # Key metrics by cluster
    metrics = ['Recency', 'Frequency', 'TotalSpend', 'AvgSpend']
    fig, axes = plt.subplots(2, 2, figsize=(15, 12))
    for i, metric in enumerate(metrics):
        ax = axes[i//2, i%2]
        sns.boxplot(data=clustered_data, x='Cluster', y=metric, ax=ax)
        ax.set_title(f'{metric} by Cluster')
    plt.tight_layout()
    plt.show()

def main():
    """Main execution function"""
    
    # Create features for clustering
    print("Creating customer features for clustering...")
    features = create_clustering_features(customers_df, transactions_df, products_df)
    
    # Find optimal number of clusters
    print("\nFinding optimal number of clusters...")
    optimal_clusters, db_scores = find_optimal_clusters(features)
    print(f"\nOptimal number of clusters: {optimal_clusters}")
    
    # Perform clustering
    print("\nPerforming clustering...")
    clustered_data, cluster_centers = perform_clustering(features, optimal_clusters)
    
    # Analyze results
    print("\nAnalyzing clusters...")
    analyze_clusters(clustered_data)
    
    # Save clustering results
    output_filename = 'FirstName_LastName_Clustering.pdf'
    # Note: You'll need to save the visualizations to the PDF
    print(f"\nResults saved to {output_filename}")
    
    return clustered_data, db_scores

# Run the clustering analysis
if __name__ == "__main__":
    clustered_data, db_scores = main()
```
